### Download Resources

Use the following syntax for downloading resources:

- **From civit.ai**: Use the command `!wget -c "link" -O "output_path/resource_name"`
  - **Example**: `!wget -c "https://civitai.com/api/download/models/125985?type=Model&format=SafeTensor" -O "./models/loras/Ghibli_v6.safetensors"`
    - This example saves the model file with the name `Ghibli_v6.safetensors` in the `./models/loras/` directory.

- **From huggingface.ai**: Use the command `!wget -c "link" -P "output_path"`
  - Replace `"link"` with the Hugging Face URL and `"output_path"` with the desired directory path.
  - **Example**: `!wget -c "https://huggingface.co/Lykon/dreamshaper-8-lcm/resolve/main/DreamShaper8_LCM.safetensors" -P "./models/checkpoints/"`
    - This example saves the file in the `./models/huggingface/` directory.

### Run ComfyUI

```python
import subprocess
import threading
import time
import socket

# Install pyngrok
!pip install pyngrok
from pyngrok import ngrok, conf

# Get ngrok authtoken from Colab secrets
# You need to add your ngrok authtoken to Colab secrets with the name NGROK_AUTH_TOKEN
# Instructions: Visit ngrok.com, create a free account, get your authtoken from the dashboard,
# then add it to Colab's secrets manager (the '🔑' icon on the left panel) with the name `NGROK_AUTH_TOKEN`.
import os

try:
    from google.colab import userdata
    NGROK_AUTH_TOKEN = userdata.get('NGROK_AUTH_TOKEN')
    if NGROK_AUTH_TOKEN:
        conf.get_default().auth_token = NGROK_AUTH_TOKEN
    else:
        print("WARNING: NGROK_AUTH_TOKEN not found in Colab secrets. ngrok tunnel might not work.")
except Exception as e:
    print(f"Could not retrieve NGROK_AUTH_TOKEN from Colab secrets: {e}")
    print("Please ensure you have added NGROK_AUTH_TOKEN to your Colab secrets.")

def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  print("\nComfyUI finished loading, trying to launch ngrok tunnel (if it gets stuck here, check your ngrok authtoken)\n")

  try:
    public_url = ngrok.connect(port).public_url
    print(f"This is the URL to access ComfyUI: {public_url}")
    print("Local ComfyUI IP: 127.0.0.1")
  except Exception as e:
    print(f"Error creating ngrok tunnel: {e}")
    print("Please ensure your NGROK_AUTH_TOKEN is correctly set in Colab secrets.")

threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```
